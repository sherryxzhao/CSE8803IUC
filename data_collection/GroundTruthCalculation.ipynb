```python
boston_data = pd.read_csv('cleaned_Boston2022_crime_data.csv')
```

```python
print(type(boston_data['OCCURRED_ON_DATE'][0]))
```

```python
boston_time = boston_data['OCCURRED_ON_DATE']
date_format = "%m/%d/%y %H:%M"

boston_crime_date = []
for date_string in boston_time:
    date = datetime.strptime(date_string, date_format)
    boston_crime_date.append(date.date())
```

```python
crime_list = list(set(boston_data['OFFENSE_DESCRIPTION'].tolist()))
```

```python
boston22 = pd.DataFrame({
    'OFFENSE_DESCRIPTION': boston_data['OFFENSE_DESCRIPTION'].tolist(),
    'OCCURRED_ON_DATE': boston_crime_date
})
```

```python
crime_counts = boston22.groupby(['OCCURRED_ON_DATE', 'OFFENSE_DESCRIPTION']).size().reset_index(name='COUNT')
print(type(crime_counts))
```

```python
pivot_table = crime_counts.pivot_table(index='OCCURRED_ON_DATE', columns='OFFENSE_DESCRIPTION', values='COUNT', aggfunc='sum')
pivot_table = pivot_table.fillna(0)
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table.loc['Total'] = pivot_table.sum()
print(pivot_table)
```

```python
pivot_table.to_csv('Boston22_groundtruth.csv', index=True)
```
